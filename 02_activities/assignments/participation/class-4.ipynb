```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import requests
import scipy
from PIL import Image
from io import BytesIO
```

```python
x = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y1 = np.array([110, 180, 240, 99, 220])
y2 = np.array([170, 100, 90, 120, 50])

fig, ax = plt.subplots(figsize=(7, 3))

ax.bar(x, y1,
color = "orange")
ax.plot(x, y2,
color = "red")
```

```python
# ErrorbarContainer object of 3 artists
y2_sd = np.std(y2)

fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2, color = "red")

ax.errorbar(x, y2, 
        yerr = y2_sd,
        fmt = "none")
```

```python
y2_sd = np.std(y2)

fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2, color = "red")

ax.errorbar(x, y2,
        yerr = y2_sd,
        fmt = "none",
        ecolor= "orange",
        elinewidth= 4,
        capsize = 6,
        capthick= 4 )
```

```python
y2_sd = np.std(y2)

fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2, color = "red")

ax.errorbar(x, y2,
        yerr = y2_sd,
        fmt = "none",
        ecolor= "orange",
        elinewidth= 4,
        capsize = 6,
        capthick= 4, 
        errorevery= 2)
```

```python
image_url = 'https://vignette.wikia.nocookie.net/onepiece/images/6/65/Tony_Tony_Chopper_Anime_Post_Timeskip_Infobox.png'
response = requests.get(image_url)
image_file = BytesIO(response.content)
image = Image.open(image_file)

```

```python
fig, ax = plt.subplots(figsize=(7,3))
ax.plot(x, y2, color='red')
# plot on top to hold image
ax_image = fig.add_axes([0.1, 0.11, 0.15, 0.35]) #x-coord, y-coord, width, height
# because we used fig., the coordinates are based off the entire figure item
# see how that little image is available
```

```python
fig, ax = plt.subplots(figsize=(7,3))
ax.plot(x, y2, color='red')
# plot on top to hold image
ax_image = fig.add_axes([0.1, 0.11, 0.15, 0.35])

ax_image.imshow(image)
ax_image.axis('off')
```
